```python
import pysrt

def extract_text_from_srt(file_path):
    subs = pysrt.open(file_path)
    text = ""
    for sub in subs:
        text += sub.text_without_tags + "\n"
    return text

# Define the path to your subtitle file
subtitle_file = "subtitles/example.srt"  # Update this with the actual path to your subtitle file

# Extract text from the subtitle file
text = extract_text_from_srt(subtitle_file)

# Display the extracted text
print(text)

# Optionally, save the extracted text to a text file
with open("extracted_text.txt", "w") as file:
    file.write(text)
```
